# Multiple Linear Regression

```python
import pandas as pd;
import numpy as np;
```

```python
dataFile = pd.read_csv('/aircraft_fuel_consumption - aircraft_fuel_consumption.csv');
dataFile
```

```python
from sklearn.model_selection import train_test_split;
from sklearn.linear_model import LinearRegression;
from sklearn.metrics import mean_squared_error, r2_score;
```

```python
# To test Impuatation i.e. for null / missing values

dataFile.isnull().sum() # zero
```

```python
# Define features and Targets;
X = dataFile[['Flight_Distance', 'Number_of_Passengers', 'Flight_Duration','Aircraft_Type']]
Y = dataFile['Fuel_Consumption']
```

```python
# Encodin Data

# convert data variable to numerical
X = pd.get_dummies(X, columns = ['Aircraft_Type'])
X
```

```python
# Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 20);
X_train.columns
```

```python
# Model

model = LinearRegression();
model.fit(X_train, Y_train);
```

```python
# Prediction

Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
# Error

from sklearn.metrics import mean_squared_error;

# Calculate MSE
mse = mean_squared_error(Y_test, Y_pred);
print(f'Mean Squared Error: {mse}');
```

```python
import joblib
# Save the model and scaler
joblib.dump(model, 'linear_regression_model.pkl')
```
